```python
doc['token'] = [nltk.word_tokenize(i) for i in doc['Report']]
doc
```

```python

unique = set(doc['Report'].str.replace('[^a-zA-Z ]', '').str.lower().str.split(' ').sum())
print(len(unique))
print(list(sorted(unique)))
token_population = []
token_population = list(sorted(unique))

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
stop_words=nltk.corpus.stopwords.words('english')

bag_of_words=[]
for w in token_population:
    if w not in stop_words:
        bag_of_words.append(w)
        
print(len(bag_of_words))
        
```

```python
def vectorize(tokens):
    vector=[]
    for w in bag_of_words:
        if w in tokens:
            vector.append(1)
        else:
            vector.append(0)
    return vector

doc['vect']=[vectorize(i) for i in doc['token']]

len(doc.iloc[: 6])
```

```python
def second(tokens):
    vec=[]
    for w in bag_of_words:
        vec.append(tokens.count(w))
    return vec
doc['vect2']=[second(i) for i in doc['token']]

```

```python
import numpy as np
templist=[]
templist=doc['vect'].tolist()
jaccard_coeff=[]

for x in range(37):
    for y in range(37):
        v1=templist[x]
        v2=templist[y]        
        intersection = np.logical_and(v1, v2)
        union = np.logical_or(v1, v2)
        similarity = intersection.sum() / float(union.sum())
        #print(' Similarity between x and y is', simxy)
        jaccard_coeff.append(similarity)

jacc=np.array(jaccard_coeff) 

#print(type(jaccard_coef))  
print(jacc.shape)
jacc=jacc.reshape(37,37)
print(jacc.shape)
print(jacc)
```

```python
import numpy as np
templist2=[]
templist2=doc['vect2'].tolist()

#print(templist2)

cos = []
import math
def cosine_similarity(v1,v2):
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)

for x in range(37):
    for y in range(37):
        v1 = templist2[x]
        v2 = templist2[y]        
        
        cos.append(cosine_similarity(v1, v2))

cosine=np.array(cos)
 
print(cosine.shape)
cosine=cosine.reshape(37,37)
print(cosine.shape)
print(cosine)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20,20))
import seaborn as sns
sns.heatmap(cosine,annot=True)
ax.set_title('Cosine Similarity')
```
