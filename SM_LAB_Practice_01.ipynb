```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create example dataframe
df = pd.read_csv('/content/dataset1.csv')

# Print original dataframe
print(df)
```

```python
""" Question 01 """

# Remove 'SNo' column
df.drop(columns=['Sno'], inplace=True)

# Print updated dataframe
print(df)
```

```python
""" Question 02 """

# Fill the missing values in Maxpulse column with 120
df['Maxpulse'].fillna(120, inplace=True)

# fill the missing values in Calories column with its mean
df['Calories'].fillna(df['Calories'].mean(), inplace=True)

print(df)
```

```python
""" Question 03 """

# Adding new row to the dataframe
new_row = {'Duration': 30, 'Date': '2020/12/11', 'Pulse': 103, 'Maxpulse': 145, 'Calories': 319.3}
df = df.append(new_row, ignore_index=True)

print(df)
```

```python
""" Question 04 """

# Sort dataframe by 'Maxpulse' in descending order
df.sort_values(by='Maxpulse', axis=0, ascending=False, inplace=True)

print(df)
```

```python
""" Question 05 """

# Create the boxplot
sns.boxplot(x="Date", y="Pulse", data=df)

# Add x-axis and y-axis labels
plt.xlabel('Date')
plt.ylabel('Pulse')

# Show the plot
plt.show()
```

```python
""" Dataset 03 """

# Create example dataframe
df = pd.read_csv('/content/dataset3.csv')

df.head()
```

```python
""" Question 01 """

# Replace missing values of total_bill with its median
median = df['total_bill'].median()
df['total_bill'] = df['total_bill'].fillna(median)

# Replace missing values of day with its mode
mode = df['day'].mode()[0]
df['day'] = df['day'].fillna(mode)
```

```python
df
```

```python
""" Question 02 """

# filter for non smokers
non_smokers = df[df['smoker'] == 'No']

# get max and min tip for non smokers
max_tip = non_smokers['tip'].max()
min_tip = non_smokers['tip'].min()

print("Maximum tip given by non smokers: ", max_tip)
print("Minimum tip given by non smokers: ", min_tip)
```

```python
""" Question 03 """

# update the day of the second row to 'Sat'
df.at[1, 'day'] = 'Sat'

print(df.head(3))
```

```python
""" Question 04 """

# create scatterplot
plt.scatter(df['total_bill'], df['tip'])

# add x and y labels
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
